# Logistic Regression :

## Step 1: Data Preprocessing & Exploration

```python
#import required libaries for data preprocessing.
import pandas as pd

data=pd.read_csv('Downloads/archive/water_potability.csv')
data.head
```

```python
data.isnull().sum()
#to check for any null or missing values.
data['ph'].isnull().sum()
data['ph'].mean()
data['ph'].fillna('7',inplace=True)
data['ph'].isnull().sum()
data['Sulfate'].isnull().sum()
data['Sulfate'].mean()
data['Sulfate'].fillna('333',inplace=True)
data['Sulfate'].isnull().sum()
data['Trihalomethanes'].isnull().sum()
data['Trihalomethanes'].mean()
data['Trihalomethanes'].fillna('66',inplace=True)
data['Trihalomethanes'].isnull().sum()
```

```python
data.isnull().sum()
data.isnull().sum().sum()
# There is no missimg value, now
```

```python
#converting data into datatype int.
prepareddata=data.astype(int)
prepareddata.head()
```

## Step 2: Data Visualization

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Visualization using pariplot 
sns.pairplot(data=prepareddata,hue='Potability')
```

## Step 3: Data Training & Model Creation

```python
# Importing train_test_split from sklearn.model_selection
from sklearn.model_selection import train_test_split


# X is the data which will have features for classification
# Y will have our target.
x=prepareddata.drop(['Potability'],axis=1)  
y=prepareddata['Potability']


# Split data into training data(80%) and testing data(20%).
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=500)


# Importing logistic regression model
from sklearn.linear_model import LogisticRegression 
LoRe = LogisticRegression()


# Fitting data into the model.
LoRe.fit(x_train, y_train)


# Making predictions 
pred = LoRe.predict(x_test)
```

```python
pred
```

## Step 4: Performance Evaluation 

```python
print("Accuracy of the Logistic Regression model is: \n ") 
print(LoRe.score(x_train,y_train))
```

#### Conclusion:

By using Logistic Regression model, the accuracy we got is 61.60 %.
